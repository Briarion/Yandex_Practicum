### Объединение таблиц

Проведем объединение таблиц в одну выборку по столбцу customerID

```python
data = contract_new.merge(personal_new, on='customerID', how='left')
data = data.merge(phone_new, on='customerID', how='left')
data = data.merge(internet_new, on='customerID', how='left')
data.head()
```

Приведем названия столбцов к нижнему регистру и разделитель между словами

```python
data.columns = data.columns.str[:-1]+data.columns.str.lower().str[-1]
data.columns = data.columns.str.upper().str[0]+data.columns.str[1:]
data.columns = data.columns.str.replace(r"([A-Z])", r"_\1").str.lower().str[1:]
data.info()
```

Преобразуем тип данных в столбце begin_date к типу datetime

```python
data['begin_date'] =pd.to_datetime(data['begin_date'])
data['begin_date'].sort_values()
```

### Целевой признак и дополнительный признаки для обучения

Формуируем целевой признак на основе столбца end_date

```python
data.loc[(data['end_date'] == 'No'), 'target'] = 0
data['target'] = data['target'].fillna(1)
data['target'] = data['target'].astype('int')
```

Заполняем поле end_date для лояльных клиентов датой выгрузки

```python
data.loc[(data['end_date'] == 'No'), 'end_date'] = '2020-02-01'
data['end_date'] =pd.to_datetime(data['end_date'])
```

Формируем дополнительный признак на основе разницы дат begin_date и end_date, оставляем только колличество полных месяцев, т.к. оплата за услуги идет ежемесячная

```python
data['month_total'] = (12 - data['begin_date'].dt.month) - (12 - data['end_date'].dt.month) + 12 * (data['end_date'].dt.year - data['begin_date'].dt.year)
```

```python
data['day_total'] = (data['end_date'] - data['begin_date']).dt.days
```

### Обработка пропусков

Преобразуем тип данных слотбца total_charges из строки к числовому

```python
data['total_charges'] = pd.to_numeric(data['total_charges'], errors='coerce')
```

Заполним пропуски в столбце произведением месячного платежа monthly_charges на общее колличество месяцев, за которые предоставлялись услуги

```python
data['total_charges'] = data['total_charges'].fillna(data['monthly_charges']*data['month_total'])
```

## Анализ данных

```python
def compare_target(data, column, title):
    res = []
    index = []
    for i in data[column].unique():
        a = round(data[column][(data['target'] == 1) & (data[column] == i)].count()/data[column][(data['target'] == 1)].count()*100, 1)
        b = round(data[column][(data['target'] == 0) & (data[column] == i)].count()/data[column][(data['target'] == 0)].count()*100, 1)
        res.append([a, b])
        index.append(i)
    res = pd.DataFrame(res, columns=['Lost', 'Loyal'], index=index)
    return res.sort_values(by='Lost', ascending=False).plot(kind='pie', subplots='True', autopct='%.2f', figsize=(12,6), title=title, legend=False);
```

```python
def model_ef (model, cv, target, feaches):
    cv_split = TimeSeriesSplit(n_splits=cv).split(features_train)
    cv_results = cross_validate(model, feaches, target, cv=cv_split, scoring=roc_auc_scorer)
    return cv_results
```

```python
cv_results = []
models = [model_forest, model_cat, model_logistic]
for i in models:
    res = model_ef(i, 5, target_train, features_train)
    if i == model_logistic:
        res = model_ef(i, 5, target_train, features_train_linear)
    cv_results.append(abs(res['test_score'].mean()))
cv_results = pd.DataFrame(cv_results, columns=['ROC AUC'], index=['RandomForest', 'CatBoost', 'LogisticRegression'])
cv_results.sort_values(by='ROC AUC', ascending=False)
```
