### Обработка пропусков и преобразование типов

Выведем строки с пустыми значениями в столбце name.

```python
data[data['name'].isna()]
```

Данные строки имеют большое колличество пропусков и не несут полезной информации, можно удалить их из датафрейма без исскажения данных будущего исследования. Сохраним данные в новый датафрейм с коорым будем проводить все дальнейшие операции и сохраним исходный датафрейм, если в дальнейшем потребуется необходимость в исходных данных.

```python
df = data.dropna(subset='name').reset_index(drop=True)
```

Обработаем пропуски в столбце user_score

```python
display(df['user_score'].unique())
print('Всего записей tbd в столбце user_score:', df['user_score'][df['user_score'] == 'tbd'].count())
print('Всего записей NaN в столбце user_score:', df['user_score'].isna().sum())
print('Процент записей tbd к общему количеству записей:', 
      round(df['user_score'][df['user_score'] == 'tbd'].count()/df['name'].count()*100, 1))
```

В столбце user_score присутствует записть tbd (To Be Determinated/Подлежит Уточнению), можно считать что в строках с данной записью данные не точные или не полные. Можно удалить их из датафрейма, но это почти 14.5% от общего колличества записей и так как это слишком большой процент от общей выборки, а заменить эти данные или уточнить нет возможности, то в данном столбце следует заменить tbd на пустое значение

```python
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')
print('Колличество записей tbd в столбце user_score', df['user_score'][df['user_score'] == 'tbd'].count())
print('Колличество записей NaN в столбце user_score', df['user_score'].isna().sum())
```

Можно заметить что помимо пропусков в годах можно заполнить пропуски в оценках, но если обратить внимание, то для одной и тойже игры оценки как критиков, так и пользователей отличаются, соответственно если так заполнить пропуски, то информация будет не достоверна.

Заполним пропуски в столбце year_of_release соответствующими значениями из строк с одинамковым названием игры.

```python
print('Колличество пустых значений в столбце year_of_release', df['year_of_release'].isna().sum())
for i in df['name'].unique():
    df.loc[(df['name'] == i) & (df['year_of_release'].isnull()), 'year_of_release'] = df.loc[
        (df['name'] == i), 'year_of_release'].min()
print('Колличество пустых значений в столбце year_of_release после заполнения', df['year_of_release'].isna().sum())
```

Оставшиеся 146 записей нет возможности достоверно заполнить и так как на основе года выпуска будет проводиться дальнейший анализ, то данные строки можно отфильтровать из датафрейма.

```python
df = df.dropna(subset='year_of_release').reset_index(drop=True)
print('Колличество пустых значений в столбце year_of_release', df['year_of_release'].isna().sum())
```

Пустые значения в столбцах user_score и critic_score можно заполнить из не пустых значений в той же строке, но сначала следует проверить кореляцию между столбцами

```python
print('Кореляция между столбцами user_score и critic_score:', round(df['user_score'].corr(df['critic_score']), 2))
#df['user_score'] = df.loc[df['critic_score'].notna(), 'user_score'].fillna(df['critic_score']/10)
#df['critic_score'] = df.loc[df['user_score'].notna(), 'critic_score'].fillna(df['user_score']*10)
```

Прямая кореляция присутствует, но не достаточно высокая что бы заполнить значения из соседнего столбца в одной строке. Заодно приведем оба столбца к одному виду, более привычному для глаза, к оценке от 1 до 10.

```python
df['critic_score'] = df['critic_score'] / 10
```

После удаления/заполнения пустых значений в столбце year_of_release можно изменить тип данных стоблца на integer.

```python
df['year_of_release'] = df['year_of_release'].astype(int)
```

Добавим в датафрейм стобец включающий в себя продажи по всем трем регионам и подправим последовательность столбцов для большей наглядности.

```python
df['total_sales'] = df['na_sales'] + df['eu_sales'] +df['jp_sales'] + df['other_sales']
df = df.reindex(columns=['name', 'platform', 'year_of_release', 'genre', 'na_sales', 'eu_sales',
       'jp_sales', 'other_sales', 'total_sales', 'critic_score', 'user_score', 'rating'])
```

Из графика видно, что до 1995 года игровая индустрия только начинала свое зарождение и начиная с 1995 получила толчек в развитии, после чего развивалась линейно до 2003, после чего снова толчек в развитии и линейный рост с небольшой просадкой в 2004 и 2005. Данные периоды имеют закономерность - пик роста, небольшой спад и бурный последующий рост. Соответственно можно сделать предположение что после падений продаж в 2013-2015 годах последует буйный рост индустрии, что подтверждают продажи в 2016 году.

Посчитаем колличество игр выпощенных до 1995, в 1995 и до 2000 и колличество не пустых записей в столбцах user_score и critic_score

Построим график продаж по игровым платформам по всему доступному временному промежутку, что бы выявить платформы с самыми высокими продажами

```python
(
df.pivot_table(index='platform', values='total_sales', aggfunc='sum')
    .sort_values(by='total_sales')
    .plot(kind='bar', grid=True, figsize=(10, 4), legend=False, xlabel='Игровые платформы', ylabel='Общие продажи')
);
plt.title('Распределение игр по платформам в зависимости от продаж');
```

Больше всего на графике выделяются 6 игровых платформ: PS2, PS3, PS, X360, DS, Wii. Построим для каждой платформы из списка график продаж распределенный по годам.

```python
platform = ['PS2', 'PS3', 'PS', 'X360', 'DS', 'Wii']
i = 0
j = 0

fig, axes = plt.subplots(nrows = 3, ncols =2 )


for n in platform:
    (
    df[df['platform'] == n]
        .pivot_table(index='year_of_release', values='total_sales', aggfunc='sum')
        .plot(kind='bar', grid=True, legend=False, rot=0, figsize=(5, 0.9), title=n, fontsize=6, 
              xlabel='', ylabel='Колличество игр', ax=axes[i, j])
    )
    j+=1
    if j > 1:
        j =0
        i+=1

fig.set_size_inches(12, 8)
plt.show()
```

Из графиков видно что цикл от выпуска до заката популярности платформы в среднем 10 лет. Из них 2-3 года на первые продажи и начало роста популярности и около 3х лет на закат популярности. Пик продаж приходитсяна 4-5 лет срока жизни игровой платформы.

Для актуального прогноза на 2017 год нам потребуются данные не более 4х летней давности. Отфильтруем в новый датафрейм записи с 2013 года и новее.

### Проверка равенства средних выборок из двух генеральных совокупностей

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Проверим гипотезу, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Зададим пороговое значение за 0.05. Нулевая гипотеза будет звучать: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

```python
alpha = .05

results1 = st.ttest_ind(
    filtered_df[(filtered_df['platform'] == 'PS4') & (filtered_df['user_score'].notna())]['user_score'], 
    filtered_df[(filtered_df['platform'] == 'XOne') & (filtered_df['user_score'].notna())]['user_score'])

print('p-значение:', results1.pvalue)

if results1.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Значение p-value больше порогового значения уровня значимости 0.05. Нулевая гипонеза не отклонена.

#### Средние пользовательские рейтинги жанров Action и Sports разные.

Проверим гипотезу, что Средние пользовательские рейтинги жанров Action и Sports разные. Зададим пороговое значение за 0.05. Нулевая гипотеза будет звучать: Средние пользовательские рейтинги жанров Action и Sports одинаковые.

```python
alpha = .05

results2 = st.ttest_ind(
    filtered_df[(filtered_df['genre'] == 'Action') & (filtered_df['user_score'].notna())]['user_score'], 
    filtered_df[(filtered_df['genre'] == 'Sports') & (filtered_df['user_score'].notna())]['user_score'])

print('p-значение:', results2.pvalue)

if results2.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Значение p-value меньше порогового значения уровня значимости 0.05. Нулевая гипонеза отклонена.

Нулевая гипотеза формируется на основе утверждения что разницы и зависимости между двумя наблюдаемыми событиями нет, опровержение нулевой гипотизы означает, что связь между двумя событиями существует.

Пороговое значение уровня значимости выбирается в диапазоне 0.01-0.05 и оно означет уровень на котором шанс получить данные о наблюдаемых событиях случайно считается допустимым. Т.е. при alpha=0.05, шанс выше, что нулевая гипотеза опровергнута основываясь на данных которые получены случайно и ошибочно, чем при alpha=0.01, где шансы на ошибочные данный меньше, но и нулевую гипотезу опровергнуть сложнее.

На основе проведенных тестов можно делать говорить о правдоподобности сделанных выводов, а не массы случайностей которые лишь выглядят как взаимосвязи.
