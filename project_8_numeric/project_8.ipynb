# Определение стоимости автомобилей

Сервис по продаже автомобилей с пробегом «Не бит, не крашен» разрабатывает приложение для привлечения новых клиентов. В нём можно быстро узнать рыночную стоимость своего автомобиля. В вашем распоряжении исторические данные: технические характеристики, комплектации и цены автомобилей. Вам нужно построить модель для определения стоимости. 

Заказчику важны:

- качество предсказания;
- скорость предсказания;
- время обучения.

______________

Перед нами стоит задача построения модели машинного обучения, которая предсказывает стоимость автомобилей с пробегом на основе технических характеристик и комплектации. Основные требования к модели машинного обучения: точность предсказания(не менее 2500 по RMSE), скорость предсказания. и время обучения модели. 

Для обучения предоставленны данные о 354,5 тысячах объявлений о продаже подержанных автомобилей скачанных из базы в 2016 году. Перед исследованием проведем подготовку предоставленных данных, проверим данные на пропуски, выбросы и проанализируем какие именно из предоставленных признаков будут полезны для модели, а какие на модель не имеют влияния.

Для иследования будем использовать 3 модели машинного обучения: CatBoostRegressor, LightGBM Regressor и RandomForesRegressor. Для каждой модели подберем не менее трех гиперпараметров с помощью кроссвалидации через GridSearch. Обученные модели с подобранными гиперпараметры сравним между собой и с Dummy моделью.

Итогом исследования будет являться модель с наилучгими показателями точности и скорости работы, либо компромисное решение на основе полученных результатов.

#### Функции для графиков

```python
def plt_hist(data, column, title, bins):
    plt.figure(figsize=(12,6))
    sns.histplot(data, x=column, bins=bins)
    sns.set(style='white')
    plt.title(title, fontsize=16)
    plt.ylabel('Колличество значений')
    if data[column].value_counts().shape[0] >20:
        plt.xticks(rotation=90)
    plt.show()
```

```python
def boxplot(data, x):
    plt.figure(figsize=(12,8));
    (
    sns.boxplot(x=x, y='price',
               data=data.sort_values(by=x))
                .set(xlabel=x, ylabel='Цена автомобилей', 
                     title='Распределение цен автомобилей по столбцу '+x, )
    );
    if 8 <data[x].value_counts().shape[0] < 12:
        plt.xticks(rotation=45)
    elif data[x].value_counts().shape[0] > 12:
        plt.xticks(rotation=90)
```

```python
def categorize(power):
    if power <= 65:
        return 65
    if 65 < power <= 80:
        return 80
    if 80 < power <= 99:  
        return 99
    if 99 < power <= 120:
        return 120
    if 120 < power <= 150:
        return 150
    if 150 < power <= 180:
        return 180
    if 180 < power <= 200:
        return 200
    if 200 < power <= 230:
        return 230
    if 230 < power <= 250:
        return 250

training_data['power'] =training_data['power'].apply(categorize)
```

## Анализ моделей

Проверим скорость обучения/предсказания для каждой модели. Для каждой модели замеряем RMSE через кросс-валидацию и  время на затраченное на обучение/предсказание для тестового набора данных и повторяем процедуру 10 раз. Запишем каждый результат в отдельную строку датафрейма.

```python
def time_ef (model, times, cv, target, feaches):
    cv_results = cross_validate(model, feaches, target, cv=cv, scoring=RMSE, return_estimator=True)
    df = []
    for i in range(times):  
        start = time.time()
        for j in range(cv):
            model.fit(feaches, target)
        end = time.time()
        training = (end-start)/cv
        RMSE_score = abs(cv_results['test_score'].mean())
        start = time.time()
        for j in range(cv):
            prediction = model.predict(feaches)
        end = time.time()
        predictions = (end-start)/cv
        df.append([training, predictions, RMSE_score])
    return pd.DataFrame(df, columns=['training', 'predictions', 'RMSE']), cv_results
```
